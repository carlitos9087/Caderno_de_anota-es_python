```python
def ordenandor_titulos(titulos_iniciais:list):
    lista_nomes_corretos=[]
    
    for titulo in titulos_iniciais:
        padrao = re.compile(r"[a-zA-Z]*[-]*[_]*[a-zA-Z]*_[0-9]{1,2}")
        resultado = padrao.findall(titulo)
        modificado = resultado[0]
        
        if len(modificado) == 14: 
            modificado = modificado.split("_")

            if len(modificado) == 2:
                modificado.insert(len(modificado)-1,"_0")
                modificado = "".join(modificado)
            
                
            elif len(modificado) == 3:
                modificado.insert(1,"_")
                modificado.insert(len(modificado)-1,"_0")
                modificado = "".join(modificado)
                
        lista_nomes_corretos.append(modificado)
    lista_nomes_corretos
    
    return lista_nomes_corretos
```

```python
def capta_tabelas(caminho:list):
    df=[]
    for num, pasta in enumerate(os.listdir(caminho)):
        df.append(pd.read_csv(os.path.join(caminho, pasta)))   #lendo tabelas tabelas
        df[num].set_index('Geneid', inplace=True)  #tornando a colula Geneid o id das tabelas
    return df 

    
```

```python
def junta_tabelas(lista_tabelas:list, lista_titulos:list):
    df_todos = pd.merge(lista_tabelas[0], lista_tabelas[1], on="Geneid")

    for itens in lista_tabelas[2:]:
        df_todos = pd.merge(df_todos, itens, on="Geneid")
        
    df_todos.columns = lista_titulos
    return df_todos
    
```

```python
def somas_reads(dataframe: list):
    somas=[[i for i in dataframe.sum()]]
    somas
    df_somas = pd.DataFrame(somas, columns=nomes)
    return df_somas/1000000
    
```
